Aplicando ferramentas de ias para prever score de credito



```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

tts = train_test_split
le = LabelEncoder()
#importa base de dados
data_base = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/data_analytics/ia_analise_de_credito/clientes.csv")
display(data_base)
```

```python
#prepara a base de dados para a IA
    #modifica os tipos das variaveis para int, para que a IA possa entendê-las

for coluna in data_base.columns:

    if data_base[coluna].dtype == "object" and coluna != "score_credito":
       data_base[coluna] = le.fit_transform(data_base[coluna])

display(data_base.info())
```

```python
#machine learning
#x são as colunas que a IA utilizará para fazer as previsões
    #Não utilizaremos as colunas id_cliente (dados aleatórios), e score_credito (dados que eu quero prever)
x = data_base.drop(columns=["id_cliente", "score_credito"])

#y é a coluna que eu quero prever

y = data_base["score_credito"]

#separar a base de dados em dados de treinos e dados de texte, tanto para x como para y

x_train, x_test, y_train, y_test = tts(x,y, test_size=0.25)
```

```python
#Cria e treina os modelos de IA
#Arvore de decisão -> RandomForest
#importa a IA
from sklearn.ensemble import RandomForestClassifier

#Cria a IA
rfc = RandomForestClassifier()

#Treina a IA
rfc.fit(x_train, y_train)


#KNN -> Nearest Neighbors
#importa a IA
from sklearn.neighbors import KNeighborsClassifier

#cria a IA
knn = KNeighborsClassifier()

#Treina a IA
knn.fit(x_train, y_train)
```

```python
#Analisar e escolher modelos de IAs
#realiza previsões na rfc
prev_rfc = rfc.predict(x_test)

##realiza previsões na knn
prev_knn = knn.predict(x_test)
```

```python
#testa e compara os modelos
#verifica acuracia
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prev_rfc))

print(accuracy_score(y_test, prev_knn))
```

```python
#Configurar o melhor modelo (o rfc) para fazer novas previsões
#importar novos clientes
new_data_base = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/data_analytics/ia_analise_de_credito/novos_clientes.csv")

#codificar novos clientes
for data in new_data_base.columns:
    if new_data_base[data].dtype == "object" and data != "score_credito":
        new_data_base[data] = le.fit_transform(new_data_base[data])

display(new_data_base.info())
```
